```python
# ====Search Path with RRT====
size = [0.5, 0.5, 0.5, 0.5]
obstacleList = [
    (5, 5, size),
    (4, 5, size),
    (4, 6, size),
    (4, 7, size),
    (4, 8, size),
    (4, 10, size),
    (6, 5, size),
    (7, 5, size),
    (8, 6, size),
    (8, 8, size),
    (8, 10, size)
]  # [x,y,size(radius)]

# Set Initial parameters
start = [0.0, 0.0, np.deg2rad(180.0)]
goal = [6.0, 7.0, np.deg2rad(90.0)]

rrt_star_reeds_shepp = RRTStarReedsShepp(goal, start,
                                            obstacleList,
                                            [-2.0, 15.0],
                                            max_iter=max_iter)
path = rrt_star_reeds_shepp.planning(animation=show_animation)

# Draw final path
if path:  # pragma: no cover
    rrt_star_reeds_shepp.draw_graph()
    plt.plot([x for (x, y, yaw) in path], [y for (x, y, yaw) in path], '-r')
    plt.grid(True)
    plt.pause(0.001)
    plt.show()
```
