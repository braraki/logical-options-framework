```python
show_animation = False
gx=8.0
gy=6.0
# ====Search Path with RRT====
obstacleList = [
    (5, 5, 1),
    (6, 5, 1),
    (7, 5, 1),
    (8, 5, 1),
    (9, 5, 1),
    (10, 5, 1),
    (5, 6, 1),
    (5, 7, 1),
    (5, 8, 1),
    (5, 9, 1),
    (5, 10, 1),
    (10, 6, 1),
    (10, 7, 1),
    (10, 8, 1),
    (6, 10, 1),
    (7, 10, 1),
    (8, 10, 1),
    (9, 10, 1),
    (8, 7, 1),
    (7, 7, 1),
    (9, 7, 1)
]  # [x, y, radius]
path = [
    (5, 4),
    (6, 4),
    (7, 4),
    (8, 4),
    (9, 4),
    (10, 4),
    (11, 4),
    (11, 5),
    (11, 6),
    (11, 7),
    (11, 8),
    (11, 9),
    (10, 9),
    (9, 9),
    (8, 9),
    (7, 9)
]
# Set Initial parameters
rrt = RRT(start=[4, 4],
            goal=[gx, gy],
            rand_area=[-2, 15],
            obstacle_list=obstacleList,
            path=path)
path = rrt.planning(animation=show_animation)

if path is None:
    print("Cannot find path")
else:
    print("found path!!")

# Path smoothing
maxIter = 1000
smoothedPath = path_smoothing(path, maxIter, obstacleList)

rrt.draw_graph(path=path, smoothed_path=smoothedPath)
plt.pause(0.01)  # Need for Mac
plt.show()
```
